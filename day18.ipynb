```python

### first practice of random forest
from sklearn.ensemble import RandomForestClassifier
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
data={

    "Hours":[1,2,3,4,5,6,7,8],
    "Attendence":[20,25,30,34,56,76,66,79],
    "pass":[0,0,0,1,1,1,1,1]
}
df=pd.DataFrame(data)
x=df[["Hours","Attendence"]]
y=df["pass"]
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)
model=RandomForestClassifier(n_estimators=100,random_state=42)
model.fit(x_train,y_train)
predictions=model.predict(x_test)
print(predictions)
accuracy=accuracy_score(y_test,predictions)
print(accuracy)
print(model.feature_importances_)
```

```python
##Employee promotion prediction
from sklearn.ensemble import RandomForestClassifier
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
data = {
    "Experience":[1,2,3,4,5,6],
    "Performance":[50,60,70,80,90,95],
    "Promoted":[0,0,0,1,1,1]
}
df=pd.DataFrame(data)
x=df[["Experience","Performance"]]
y=df["Promoted"]
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)
model=RandomForestClassifier(n_estimators=100,random_state=42)
model.fit(x_train,y_train)
predictions=model.predict(x_test)
print(predictions)
print(x_test)
newdata=[["17","90"]]
prediction1=model.predict(newdata)
print(prediction1)
print(accuracy_score(y_test,predictions))
print(model.feature_importances_)
```

```python
###customer purchase prediction

from sklearn.ensemble import RandomForestClassifier
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,classification_report
import matplotlib.pyplot as plt
data = {
    
    "Age":[20,25,30,35,40],
    "Income":[20000,30000,45000,55000,70000],
    "Buy":[0,0,1,1,1]
}
df=pd.DataFrame(data)
x=df[["Age","Income"]]
y=df["Buy"]
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)
model=RandomForestClassifier(n_estimators=100,random_state=42)
model.fit(x_train,y_train)
predictions=model.predict(x_test)
print(predictions)
print(x_test)
newdata=[["40","80000"]]
prediction1=model.predict(newdata)
print(prediction1)
print(accuracy_score(y_test,predictions))
print(model.feature_importances_)
print(classification_report(y_test,predictions))

```
