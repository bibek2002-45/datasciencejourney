```python
###navie bayes
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import pandas as pd
data={

"Hours":[1,2,3,4,5,6,7,8],

"Pass":[0,0,0,1,1,1,1,1]

}
df=pd.DataFrame(data)
x=df[["Hours"]]
y=df["Pass"]
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
model=GaussianNB()
model.fit(x_train,y_train)
prediction=model.predict(x_test)
print(prediction)
print(x_test)
print(accuracy_score(y_test,prediction))
```

```python
### disease prediction using navie bayes 
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import LabelEncoder
from sklearn.naive_bayes import BernoulliNB
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
import pandas as pd
data={
    "fever":["yes","no","yes","yes","no","yes","no","yes"],
    "cough":["yes","no","no","yes","yes","no","no","yes"],
    "headache":["yes","yes","yes","no","yes","yes","no","yes"],
    "disease":["flu","migrane","Dengue","flu","migrane","Dengue","migrane","flu"]


}
df=pd.DataFrame(data)
print(df)
x=df[["fever","cough","headache"]]
encoder=OrdinalEncoder()
x=encoder.fit_transform(x)
encodes=LabelEncoder()
y=encodes.fit_transform(df["disease"])
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
model=BernoulliNB()
model.fit(x_train,y_train)
predictions=model.predict(x_test)
print(predictions)
print(x_test)
print(accuracy_score(y_test,predictions))
```

```python
###Real dataset(iris dataset classification)
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import pandas as pd
from sklearn.datasets import load_iris
iris=load_iris()
x=iris.data
y=iris.target
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
model=GaussianNB()
model.fit(x_train,y_train)
prediction=model.predict(x_test)
print(prediction)
print(x_test)
print(accuracy_score(y_test,prediction))
```

```python
#spam email detection
from sklearn.naive_bayes import MultinomialNB
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
import pandas as pd 
df=pd.read_csv(r"C:\Users\user\Downloads\emails.csv\emails.csv")
print(df.columns)
print(df.info())
x=df["text"]
y=df["spam"]
vectorizer = CountVectorizer()

x = vectorizer.fit_transform(x)
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
model=MultinomialNB()
model.fit(x_train,y_train)
predictions=model.predict(x_test)
print(predictions)
print(accuracy_score(y_test,predictions))
```
