```python
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
import pandas as pd
data = {

"X":[1,2,2,8,8,25],

"Y":[2,2,3,8,9,80]

}
df=pd.DataFrame(data)
plt.scatter(df["X"],df["Y"])
plt.show()
model=DBSCAN(eps=3,min_samples=2)
model.fit(df)
clusters=model.labels_
df["clusters"]=clusters
print(df)
plt.scatter(df["X"],df["Y"],c=clusters,s=50)
plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
import pandas as pd
df=pd.read_csv(r"C:\Users\user\Downloads\taxi_latlong_dbscan.csv")
plt.scatter(df["latitude"],df["longitude"],s=50)
plt.show()
model=DBSCAN(min_samples=3,eps=0.001)
model.fit(df)
clusters=model.labels_
df["clusters"]=clusters
print(df)
plt.scatter(df["latitude"],df["longitude"],c=clusters,s=50)
plt.show()
print(df["clusters"].value_counts())
outliers=df[df["clusters"]==-1]
print(outliers)
print("number of outliers:",len(outliers))

```
